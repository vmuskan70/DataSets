```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
type_count=df["type"].value_counts()
type_count
```

```python
type_count.plot(kind ="bar")
plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Content Type")
plt.ylabel("Number of Title")
plt.show()
```

```python
rating_count=df["rating"].value_counts().head(10)
rating_count
```

```python
rating_count.plot(kind="bar")
plt.title("Top 10 Ratings on Netflix")
plt.xlabel("Rating")
plt.ylabel("Number of Title")
plt.show()
```

```python
year_count=df["release_year"].value_counts().sort_index()
year_count=year_count.tail(20)
```

```python
year_count.plot(kind="line")
plt.title("Netflix Content Release Trend By Year")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
df["country"]=df["country"].fillna("Unnknown")
```

```python
country_count=df["country"].value_counts().head(10)
country_count
```

```python
country_count.plot(kind="barh")
plt.title("Top 10 countries by Netflix Content")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()
```

```python
india_df=df[df["country"].fillna("").str.contains("india",case=False)]
india_df.head()
```

```python
india_type__count=india_df["type"].value_counts()
india_type__count
```

```python
india_type__count.plot(kind="bar")
plt.title("Indian Content : Movies vs TV Shows")
plt.xlabel("Content Type")
plt.ylabel("Number of Titles")
plt.show()
```
